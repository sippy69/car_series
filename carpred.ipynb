```python
i=df.drop(['car_ID','symboling','CarName','doornumber','enginelocation','drivewheel','aspiration'],axis='columns')
i
```

```python
target=df.price
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
i['fuelsystem_n']=le.fit_transform(i['fuelsystem'])

i['cylindernumber_n']=le.fit_transform(i['cylindernumber'])

i['enginetype_n']=le.fit_transform(i['enginetype'])
i['carbody_n']=le.fit_transform(i['carbody'])
i['fueltype_n']=le.fit_transform(i['fueltype'])
i

```

```python
i_n=i.drop(['fueltype','cylindernumber','enginetype','carbody','fuelsystem','price'],axis='columns')
i_n
```

```python
model=linear_model.LinearRegression()
model.fit(i_n,target)
```
